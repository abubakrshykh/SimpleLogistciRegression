```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
```

```python
#Load the data from scikit 
iris = load_iris()
dataset = pd.DataFrame(data=iris.data, columns=iris.feature_names)
dataset['target'] = iris.target
print(dataset.head())
```

```python
#Assigning the dependent and independent variables 
X =dataset.iloc[:,:-1].values
y =dataset['target']
```

```python
#spliting the data into training and test subsets.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=1)
```

```python
#Training the model an passing the dat into it.
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
model.fit(X_train,y_train)
```

```python
prediction = model.predict(X_test)
prediction
```

```python
#Checking the accuracy
accuracy =model.score(X_test, y_test)
print("Model's Accuracy is: ", accuracy)
```

```python
#Predicting the model manually
prediction = model.predict([[1, 9, 2, 7]])
print(prediction)
```

```python
#Classifying flowers into their respective classes
flower = np.array(['setosa', 'versicolor', 'virginica'])
if prediction == 2:
  print("Model Predicted that flower is: ", flower[2])
elif prediction == 1:
  print("Model Predicted that flower is: ", flower[1])
elif prediction == 0:
  print("Model Predicted that flower is: ", flower[0])
```
